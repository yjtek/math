- Definitions
    - Random variables are associated with uncertain outcome, vs deterministic in algebra
        - e.g. X ~ N(0, 1) can take on any value, but X = 2+1 is a fixed value
        - X can be discrete or continuous

    - Discrete distributions
        - Where values can only take ints, not floats
    
        - Probability Mass Function (PMF)
            - $Pr(X = x)$
            - Used for discrete random variables
                - e.g. number of heads in 10 coin flips

    - Continuous distributions
        - Imagine you are estimating the time between events, X
        - Since time can be any value (1.1, 1.01, 1.001, etc.), the $Pr(X=x) = 0$ for all values of x
        - So in the case of continuous values, instead of asking for $Pr(X=x)$, we ask for $Pr(x_1 >= X >= x_2)$ instead
        
        - Probability Density Function (PDF)
            - Since $Pr(X=x) = 0$ for all x, we cannot discuss probabilities for instantaneous values in the PMF
            - Instead, we use a **PDF**, which is denoted by $f_{X}X$
            - To discuss probabilities in a PDF, we talk about $Pr(a < X < b)$ = area under $f_{X}X$ between $a$ and $b$
            - A valid PDF $f_{X}X$ must fulfill 3 criteria
                - Defined for all values of $X$
                - $f_{X}X \ge 0$
                - Area under $f_{X}X$ is 1

    - Cumulative Distribution Function (CDF)
        - Integration of the PMF or PDF
        - Each point on the CDF represents the integral from $-\inf$ to $X=x$
        - A valid CDF $F_XX$ must fulfill 3 criteria
            - $0 \le F_XX \le 1$
            - Left endpoint is 0
            - Right endpoint is 0
            - Never decreases

- Binomial distribution
    - I toss a coin 5 times, let X be number heads can I get
        - Drawing the probability of X=1, X=2, etc. gives me the binomial distribution
    - Probability that in 5 flips, I get 2 heads?
        - $\frac{5!}{2! \cdot (5-2)!} = \binom{5}{2}$ <-- **binomial coefficient**
        - In general, $\binom{n}{k}$ counts the number of ways you can get $k$ positive outcomes from $n$ trials

    - General PMF: Number of heads in $n$ coin tosses, where Pr(Head) = $p$
        - $\binom{n}{k} p^k (1-p)^{n-k}$

    - Number of ways to arrange $n$ elements, where $k$ of them are of one type, and $n-k$ are of another
        - $n!$: Total number of ways to arrange elements (n options for first element, n-1 for second, etc.)
        - $k!$: Total number of ways to rearrange k for a given combination since they are the same (k choices for first, k-1 for second etc)
        - $(n-k)!$: Total number of ways to rearrange not k for a given combination since they are the same (k choices for first, k-1 for second etc)

- Bernoulli Distribution: Single binomial trial
    - In coin flipping, either heads or trails. So X ~ Bernoulli(0.5)
    - In dice roll, 1 or not 1. So X ~ Bernoulli(1/6)

- Uniform Distribution
    - PDF must be constant for all values within an interval (all sub-intervals equally likely)
    - $
        f_XX = \begin{cases}
            \frac{1}{b-a} & a \lt x \lt b \\
            0 & x \notin (a, b) \\
            1 & \text{if } x \ge b
        \end{cases}
    $

- Normal (Gaussian) Distribution
    - $
        \frac{1}{\sqrt{2 \cdot \pi \cdot \sigma^2}} \cdot e^{-(\frac{x - \mu}{\sigma})^2}
    $
    - $\mu$: Center of distribution
    - $\sigma$: Spread of distribution around center
    - Standardisation 
        - $Z = \frac{X- \mu}{\sigma}$ 

- Chi Squared Distribution
    - Imagine you are transmitting a signal from one source to a receiver. Due to interference, barriers etc, the signal is received with some noise
    - Let the noise be represented by Gaussian distribution $Z$
    - The **power** of the noise $W$ is represented $Z^2$, or $W = Z^2$
    - CDF of $W$
        - $\begin{align}
                F_W(w) &= Pr(W \le w) \\
                &= Pr(Z^2 \le w) \\
                &= Pr(\left | Z \right | \le \sqrt{w}) \\
                &= Pr(-\sqrt{w} \le Z \le \sqrt{w})
            \end{align}$
        
        - Visually, as $w$ increases, the CDF of W is simply the area under the Gaussian PDF expanding around the mid point
            - 
            - 
        
        - Notice that the PDF of chi-squred distribution increases faster at first, then slows down

        - The CDF of the chi-squared distribution is simply the derivative of the PDF, and so it is larger at the start, then becomes smaller

        - This is the Chi-squared distribution with **one degree of freedom**

    - Of course, $W$ could be more complicated than $Z^2$
        - $W = Z_{1}^{2} + Z_{2}^{2} + ... Z_{k}^{2}$
        - This is a chi-square distributino with $k$ degrees of freedom
