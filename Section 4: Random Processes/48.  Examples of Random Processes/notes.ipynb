### Example 48.1 (Random Amplitude Process)

- Let $A$ be a random variable, and $f$ be a constant
- We define a **random amplitude process** to be $X(t) = A \cos(2 \pi ft)$
- Once the value of $A$ is instantiated, the random amplitude process $X(t)$ is completely specified for all time $t$
- Let's simulate for $A \sim \text{Binomial}(n=5, p=0.5)$

```python
math.cos(2 * np.pi)
xvals = np.arange(-10,10,0.1)
# math.cos(xvals)
```

```python
A = np.arange(1,6,1)
f = 1
xvals = np.arange(-3,3,0.1)

for a in A:
    plt.plot(xvals, a * np.cos(2 * np.pi * f * xvals), alpha=0.3)
```

### Example 48.2 (Moving Average (MA) Process)

- Let $\{ Z[n] \}$ be a white noise process. Then, a moving average process (of order 1) $\{ X[n] \}$ is a discrete-time process defined by $$X[n] = b_0Z[n] + b_1 Z[n-1]$$

- Notice that a moving average process, as the name suggests, simply takes a weighted average of the new random variable $Z[n]$ and the previous random variable $Z[n-1]$
    - For simplicity, let's suppose $b_0 = b_1 = 0.5$
    - $$\begin{align}
        X[n] &= 0.5 Z[n] + 0.5 Z[n-1] 
        \end{align}$$

- Simulating, it should be clear that the blue is lower than green generally. This is because of the averaging effect of the MA term

```python
MA_TERMS = 2
SIZE = 30
INSTANTIATIONS = 30

zvals = [np.random.normal(0,1,size=SIZE) for _  in range(INSTANTIATIONS)]
weights = np.repeat(1.0, MA_TERMS) / MA_TERMS
ma = [np.insert(np.convolve(z, weights, mode='valid'), 0, z[0]) for z in zvals]

for i in range(INSTANTIATIONS):
    plt.plot(np.arange(0,len(ma[i]),1), ma[i], alpha=0.3, color='blue')
    plt.plot(np.arange(0,len(zvals[i]),1), zvals[i], alpha=0.3, color='green')
```
