### 1. Alice and Bob meet for lunch every day at a random time between noon and 1 P.M. Bob always arrives after Alice, but otherwise, they are equally likely to arrive at any time The joint p.d.f. of X, the time Alice arrives (in minutes after 12 P.M.), and Y, the time Bob arrives (in minutes after 12 P.M.), is given by $f(x,y) = \begin{Bmatrix} c & 0 \le x \lt y \le 60 \\ 0 & \text{otherwise}\end{Bmatrix}$, where c is constant

#### a. Sketch a birds-eye view of the joint p.d.f. Because the p.d.f. is constant, you should be able to calculate all probabilities using geometry, rather than integration.

```python
xvals = np.arange(0,60,0.1)
yvals = np.arange(0,60,0.1)
plt.xlim(-10, 70)
plt.ylim(-10, 70)
plt.fill_between(xvals, 0, 60, alpha=0.2)
```

#### b. Determine the value of c that makes this a valid joint p.d.f.

- $\int f(x,y) = c * 60 * 60 = 1$ for valid pdf
- $c = \frac{1}{3600}$

#### c. Calculate the probability that Bob arrives more than 25% later than Alice. That is, what is P(Y > 1.25X)? (Sketch a picture of this region.)

- When $Y=60$, $X = 60/1.25 = 48$
$$\begin{align}
    P(Y > 1.25X) &= \frac{0.5 * 48 * 60}{60^2} \\
    &= 0.4
    \end{align}$$

```python
xvals = np.arange(0,60,0.1)
y_125 = 1.25 * xvals
y_125 = np.where(y_125 > 60, 60, y_125)
plt.xlim(-10, 70)
plt.ylim(-10, 70)
plt.plot(xvals, y_125)
plt.fill_between(xvals, 0, 60, alpha=0.2)
plt.fill_between(xvals, y_125, 60, alpha=0.2, color='black')
```

### 3. An ecologist selects a point inside a circular sampling region according to a uniform distribution. Let X be the x-coordinate of the point selected and Y be the y-coordinate of the point selected. If the circle is centered at (0,0) and has radius r, then the joint pdf of X and Y is $f(x,y) = \begin{Bmatrix} c & x^2 + y^2 \le r^2 \\ 0 & \text{otherwise} \end{Bmatrix}$

#### a. Determine the value of c that makes this a valid joint p.d.f.

- For valid PDF, $\int f(x,y) = \pi r^2 \cdot c = 1$
- Hence, $c = \frac{1}{\pi r^2}$

#### b. What is the probability that the selected point is within r/2 of the center of the circular region? (Hint: Use geometry.)

- Area of full circle is $\pi r^2$
- Area of smaller circle is $\pi (\frac{r}{2})^2 = \frac{\pi r^2}{4}$
- Probability that selected point is within $\frac{r}{2}$ of center is $\frac{1}{4}$

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
fig.set_figheight(4)
fig.set_figwidth(4)
ax.add_artist(plt.Circle((0,0),5, alpha=0.4))
ax.add_artist(plt.Circle((0,0),2.5, alpha=0.4))
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
```

#### c. What is the probability that both X and Y differ from 0 by at most r/2?

$$\frac{r^2}{\pi r^2} = \frac{1}{\pi}$$

### 4. A company produces cans of mixed nuts containing almonds, cashews, and peanuts. Each can is exactly 1 lb, but the amount of each type of nut is random. The joint p.d.f. of X, the amount of almonds, and Y, the amount of cashews, is $f(x,y) = \begin{Bmatrix} 24xy & 0 \le x \le 1, 0 \le y \le 1, x+y \le 1 \\ 0 & \text{otherwise} \end{Bmatrix}$. Show that the probability there are more almonds than cashews is 0.5

- Since each can MUST be 1lb, y is not independent of x. In fact, y = 1-x for every value of X
- As such, the PDF can be represented as a line 
- For each value of X, there is 1 and only 1 value of Y
- $X \gt Y$ only when $X \gt 0.5$
- Since $X$ is uniform from 0 to 1, $P(X \gt 0.5) = 1 - P(X \lt 0.5) = 1 - CDF(0.5) = 1 - \frac{0.5-0}{1-0} = 0.5 = P(X \lt 0.5)$

```python
xvals = np.arange(0,1,0.01)
yvals = 1 - xvals
plt.plot(xvals, yvals)
```
