## Motivating Example

- Imagine a Texas holdem game with 2 players
    - Player A has QD and 3D (queen diamond, 3 diamond)
    - Player B has JC and JH (jack club, jack heart)

- They are waiting for the river cards
    - This must be one of $\binom{48}{5}$ possibilities

- Player A is interested in playing for a flush, so is interested in the number of diamonds in the river. Let X be the random variable associated with this.

- Player B is interested in playing for a 4 of a kind, so is interested in the number of Jacks in the river. Let Y be the random variable associated with this.

- How much should players A and B bet to see the river drawn?

## Theory

- A random variable is a variable that can take a range of values with some associated probabilities. Until it is observed, you do not know with certainty what value it will take

- Definition 10.1: The possible values taken by a random variable X and its associated probabilities are given by its probability mass function (PMF). This is also called the distribution of X
    - $f(x) = P(X = x)$

### Solving motivating example

- Example 10.1: From the motivating example, let's assume you are player A. What is the probability that you get a flush?
    - There are $\binom{48}{5}$ possible combinations for the river
    - In the 48 remaining cards, 11 are diamonds, and 37 are not diamonds (here, we assume perfect knowledge of your opponent's hand)
    - There can be anywhere between 0 and 5 diamonds in the river
        - 0D: $\binom{37}{5} * \binom{11}{0}$
        - 1D: $\binom{37}{4} * \binom{11}{1}$
        - 2D: $\binom{37}{3} * \binom{11}{2}$
        - 3D: $\binom{37}{2} * \binom{11}{3}$
        - 4D: $\binom{37}{1} * \binom{11}{4}$
        - 5D: $\binom{37}{0} * \binom{11}{5}$
    - For a flush, at least 3 more diamonds are needed
    - So probability of flush is: $\frac{\binom{37}{2} * \binom{11}{3} + \binom{37}{1} * \binom{11}{4} + \binom{37}{0} * \binom{11}{5}}{\binom{48}{5}} = 0.0715$

```python
from math import comb
(comb(37,2)*comb(11,3) + comb(37,1)*comb(11,4) + comb(37,0)*comb(11,5))/(comb(48,5))
```

- Example 10.2: 
    - While this is not too helpful in this example, note that the probability of flush can also be computed using the complement rule $1 - (\frac{\binom{37}{3} * \binom{11}{2} + \binom{37}{4} * \binom{11}{1} + \binom{37}{5} * \binom{11}{0}}{\binom{48}{5}}) = 0.0715$
