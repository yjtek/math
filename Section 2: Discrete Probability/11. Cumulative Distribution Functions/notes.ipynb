## Theory

- We covered Probability Mass Function (PMF) last section
    - That gives the probability of each possible outcome
    - Recall $f(X) = P(X=x)$

- Definition 11.1: Another way to visualise probability of outcomes is to find the cumulative probability of the PMF, which is known as the cumulative distribution function CDF
    - $F(X) = P(X \leq x)$

- Example 11.1: Using the same motivating example in section 10, let's compute the CDF of each number of diamonds in the community cards
    - $F(X <= 0) = \binom{37}{5} * \binom{11}{0}$
    - $F(X <= 1) = \binom{37}{5} * \binom{11}{0} + \binom{37}{4} * \binom{11}{1}$
    - ...
    - There is no general formula, so we describe CDF piecewise:

    - $$
        F(X) = \left \{ \begin{matrix}
                0 & x \leq 0 \\ 
                0.255 & 0 \leq x \leq 1 \\ 
                0.679 & 1 \leq x \leq 2 \\ 
                0.928 & 2 \leq x \leq 3 \\ 
                0.993 & 3 \leq x \leq 4 \\ 
                0.999 & 4 \leq x \leq 5 \\ 
                1.0 & x \geq 5 
        \end{matrix}\right.
    $$

    - Notes
        - Since outcomes are discrete here, the CDF is constant between integers
        - F(X) increases as x increases

- Example 11.3: Using the CDF
    - With the CDF, some probabilities are much easier to compute. For instance, probability of diamond flush is simply $1 - F(2)$ because you need at least 3 diamonds for flush

- Note:
    - You can always compute PDF from CDF
    - For example: 

$$\begin{align}
    P(X=2) &= f(3) \\
    &= P(X \leq 3) - P(X \leq 2) \\
    &= F(3) - F(2) \\
    &= 0.064
\end{align}$$
