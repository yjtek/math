## Theory

- A random walk is a good example of how a random process evolves over time

- Think of a random walk in this way
    - A person is coming home drunk
    - At any point, he can either take 1 step forward, or one step back
    - Every step is independent of every other step

- Let $Z_i$ be the random variable denoting the step at each point in time 
    - Z is either 1 with probability 0.5, or -1 with probability 0.5

- Let $X$ be the position of the person after $n$ steps
    - $X = Z_1 + Z_2 + ... Z_n$

- Given this, set up, note that there is pretty much no constraint to where the drunkard can end up

```python
import numpy as np
import matplotlib.pyplot as plt
population=[-1,1]

for _ in range(10):
    random_walk=np.cumsum(np.random.choice(population, 10_000, replace=True))
    plt.plot(random_walk)
```
