- Row echelon form 
    - 1s and 0s in the diagonal
    - Anything below is 0
    - Anything above can be anything
    - Matrix is singular if diagonal contains 0s

- Rank of a matrix is number of rows - dimension of solution space
    - In a 2x2 matrix, there are 2 rows
    - If the matrix is non-singular (i.e. the 2 lines intersect at a point), solution space has dimension 0, so rank is 2
    - If matrix is singular because the lines are the same (i.e. the 2 lines intersect throughout), the solution space has dimension 1, so rank is 1
    - if matrix is singular because lines don't intersect, the solution space has dimension 2, so rank is 0
    - By definition only full rank matrices are singular

- A matrix can also be thought of as a transformation of basis vectors. A single matrix transforms each of the basis vectors of a space to some new vector, and these collectively form the basis of the new space
    - 

- Since a matrix is simply some linear transformation of basis vectors, it stands to reason that there must be some matrix to "undo" the change that the matrix makes
    - This is the inverse of the matrix

- When does a matrix not have an inverse? When the original transformation reduces the dimensionality of the space spanned by the basis vectors! Then there is no "undoing" the transformation, because no information about the original basis vectors is preserved
