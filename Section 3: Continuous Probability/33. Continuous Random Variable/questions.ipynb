### 1. Packets arrive at a certain node on the universityâ€™s intranet at 10 packets per minute, on average. Assume packet arrivals meet the assumptions of a Poisson process.

#### a. What is the pdf of T, the time of the first arrival?

- Let's start by defining a CDF. Like the motivating example in this question, note that the probability of first particle appearing by time $t$ is the same as the probability that 0 particle appears in first $t$ periods. Formally:
    - $$\begin{align}
        F(t) &= P(T < t) \\
        &= 1 - P(T >= t) \\
        &= 1 - P(X_{t} = 0) \\
        &= 1 - e^{-\lambda t} \frac{(\lambda t)^0}{0!} \\
        &= 1 - e^{-\lambda t}
        \end{align}$$

- Taking a derivative
    - $$\begin{align}
        f(t) &= F'(t) \\
        &= \lambda e^{-\lambda t}
        \end{align}$$

- Since $\lambda = 10$, $f(t) = 10 e^{-10t}$

#### b. What is the probability that the first arrival happens between 10 seconds and 30 seconds? (Note that the rate in the problem is given in minutes.)

- From results in motivating example

$$\begin{align}
    P(10 < t < 30) &= e^{-10(\frac{1}{6})} - e^{-10(\frac{1}{2})} \\
    &= 0.1821
\end{align}$$

### 3. The distance (in hundreds of miles) driven by a trucker in one day is a continuous random variable X whose cumulative distribution function (c.d.f.) is given by: $F(x) = \begin{Bmatrix} 0 & x \lt 0 \\ \frac{x^3}{216} & 0 \le x \lt 6 \\ 1 & x \gt 6 \end{Bmatrix}$

    &= 0.421
\end{align}$$

### 4. Let T be how late that a professor lets her class out (in minutes). (If T is negative, then she finishes the class early.) The p.d.f. of T is $f(t) = \begin{Bmatrix} c(2+t) & -2 \lt t \lt 0 \\ c(2-t) & 0 \le t \lt 2 \\ 0 & \text{otherwise} \end{Bmatrix}$ where $c$ is a constant

#### a. Determine the value of $c$ necessary to make this a proper p.d.f. (Hint: What does the total probability have to be?) Sketch the p.d.f.


```python
import numpy as np
import matplotlib.pyplot as plt

base = np.arange(-3, 3, 0.01)
pdf = np.where((-2 < base) & (base < 0), 1 + base/2, np.where((0 <= base) & (base < 2), 1 - base/2, 0))
plt.plot(base, pdf)
```

- $f(t) = \begin{Bmatrix} c(2+t) & -2 \lt t \lt 0 \\ c(2-t) & 0 \le t \lt 2 \\ 0 & \text{otherwise} \end{Bmatrix}$

- $F(t) = \int_{-inf}^{inf} f(t) = \begin{Bmatrix} 0 & t \le -2 \\ 2ct + \frac{ct^2}{2} & -2 \lt t \lt 0 \\ 2ct - \frac{ct^2}{2} & 0 \le t \lt 2 \\ 1 & t \ge 2 \end{Bmatrix}$

- For $f(t)$ to be proper, $\int_{-\inf}^{\inf} f(t) dt = 1$ 
$$\begin{align}
    [2ct + \frac{ct^2}{2}]_{-2}^{0} + [2ct - \frac{ct^2}{2}]_{0}^{2} &= [ (0 + 0) - (-4c + 2c) ] + [ (4c - 2c) - (0 - 0) ] \\
    &= (4c - 2c) + (4c - 2c) \\
    &= 4c \\
    &= 1
\end{align}$$

- Therefore, $c = 0.25$. We can also see this from the plotted curve above. For the triangle to enclose an area of 1, c must be 0.25

#### b. Calculate the probability that the professor finishes between 1 minute early and 30 seconds late, i.e., $P(-1 \lt t \lt 0.5)$

$$\begin{align}
    P(-1 \lt t \lt 0.5) &= [2ct + \frac{ct^2}{2}]_{-1}^{0} + [2ct - \frac{ct^2}{2}]_{0}^{0.5} \\
    &= [ (0+0) - (-2c + \frac{c}{2})] + [ (c - \frac{0.25c}{2}) - (0+0) ] \\
    &= 2c - \frac{c}{2} + c - \frac{0.25c}{2} \\
    &= 0.5 - 0.125 + 0.5 - 0.125 \\
    &= 0.75
\end{align}$$
